```python
# Data handling
import pandas as pd
import numpy as np

# Read the csv files
df = pd.read_csv('stock_market.csv')
df.head()
```

```python
condition = (df['Name'] == 'HANG SENG') | (df['Name'] == 'NIKKEI') | (df['Name'] == 'NASDAQ')
data = df[condition]
data = data.loc[:, ['Date', 'Adj Close', 'Name', 'Day_Perc_Change']]
data = data.sort_values(['Adj Close','Date'])
data.head()
```

```python
# Bokeh libraries
from bokeh.plotting import figure, show
from bokeh.io import output_file
from bokeh.models import ColumnDataSource

# Output to file
output_notebook()

# Isolate the data for the Rockets and Warriors
hangseng_data = data[data['Name'] == 'HANG SENG']
nikkei_data = data[data['Name'] == 'NIKKEI']
nasdaq_data = data[data['Name'] == 'NASDAQ']

# Create a ColumnDataSource object for each team
hangseng_cds = ColumnDataSource(hangseng_data)
nikkei_cds = ColumnDataSource(nikkei_data)
nasdaq_cds = ColumnDataSource(nasdaq_data)

# Create and configure the figure
fig_adjj = figure(x_axis_type='auto',
             plot_height=300, plot_width=600,
             title='Adj Close Market - DEFAULT TOOLBAR',
             x_axis_label='Adj Close', y_axis_label='Day Perc Change')

# Render the race as step lines
fig_adjj.step('Adj Close', 'Day_Perc_Change', 
         color='#CE1141', legend_label='hangseng', 
         source=hangseng_cds)
fig_adjj.step('Adj Close', 'Day_Perc_Change', 
         color='#006BB6', legend_label='nikkei', 
         source=nikkei_cds)
fig_adjj.step('Adj Close', 'Day_Perc_Change', 
         color='#FFD700', legend_label='nasdaq', 
         source=nasdaq_cds)

# Move the legend to the upper left corner
fig_adjj.legend.location = 'top_right'

# Show the plot
show(fig_adjj)
```

```python
# Bokeh Libraries
from bokeh.plotting import figure, show
from bokeh.io import output_file
from bokeh.models import ColumnDataSource, NumeralTickFormatter

# Output to file

output_notebook()

# Store the data in a ColumnDataSource
data1_cds = ColumnDataSource(data1)

# Specify the selection tools to be made available
select_tools = ['box_select', 'lasso_select', 'poly_select', 'tap', 'reset']

# Create the figure
fig_adj = figure(plot_height=400,
             plot_width=600,
             x_axis_label='Adj Close',
             y_axis_label='Day Perc Change',
             title='Volume',
             toolbar_location='below',
             tools=select_tools)

# Format the y-axis tick labels as percentages
fig_adj.yaxis[0].formatter = NumeralTickFormatter(format='00.0%')

# Add square representing each player
fig_adj.square(x='Adj Close',
           y='Day_Perc_Change',
           source=data1_cds,
           color='royalblue',
           selection_color='deepskyblue',
           nonselection_color='lightgray',
           nonselection_alpha=0.3)

# Visualize
show(fig_adj)
```

```python
from bokeh.plotting import figure, show
from bokeh.io import output_file
from bokeh.models import ColumnDataSource, CDSView, GroupFilter
from bokeh.layouts import row

# Output inline in the notebook
output_notebook

# Store the data in a ColumnDataSource
data2 = ColumnDataSource(data1)

# Create a view for each player
hanseng_filters = [GroupFilter(column_name='Name', group='HANG SENG')]
hanseng_view = CDSView(source=data2,
                      filters=hanseng_filters)
nikkei_filters = [GroupFilter(column_name='Name', group='NIKKEI')]
nikkei_view = CDSView(source=data2,
                      filters=nikkei_filters)
nasdaq_filters = [GroupFilter(column_name='Name', group='NASDAQ')]
nasdaq_view = CDSView(source=data2,
                      filters=nasdaq_filters)
```

```python
common_figure_kwargs = {
    'plot_width': 400,
    'x_axis_label': 'Adj Close',
    'toolbar_location': None,
}
common_circle_kwargs = {
    'x': 'Adj Close',
    'y': 'Day_Perc_Change',
    'source': data2,
    'size': 12,
    'alpha': 0.7,
}
common_hangseng_kwargs = {
    'view': hangseng_view,
    'color': '#002859',
    'legend': 'Hangseng'
}
common_nikkei_kwargs = {
    'view': nikkei_view,
    'color': '#FFC324',
    'legend': 'Nikkei'
}
common_nasdaq_kwargs = {
    'view': nasdaq_view,
    'color': '#808080',
    'legend': 'Nasdaq'
}
```

```python
hide_fig = figure(**common_figure_kwargs,
                  title='Click Legend to HIDE Data', 
                  y_axis_label='Day Perc Change')
hide_fig.circle(**common_circle_kwargs, **common_hangseng_kwargs)
hide_fig.circle(**common_circle_kwargs, **common_nikkei_kwargs)
hide_fig.circle(**common_circle_kwargs, **common_nasdaq_kwargs)

mute_fig = figure(**common_figure_kwargs, title='Click Legend to MUTE Data')
mute_fig.circle(**common_circle_kwargs, **common_hangseng_kwargs,
                muted_alpha=0.1)
mute_fig.circle(**common_circle_kwargs, **common_nikkei_kwargs,
                muted_alpha=0.1)
mute_fig.circle(**common_circle_kwargs, **common_nasdaq_kwargs,
                muted_alpha=0.1)
```

```python
# Add interactivity to the legend
hide_fig.legend.click_policy = 'hide'
mute_fig.legend.click_policy = 'mute'

# Visualize
show(row(hide_fig, mute_fig))
```
